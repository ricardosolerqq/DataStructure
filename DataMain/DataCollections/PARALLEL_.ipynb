## runNt

```
#executa um notebook parassando um parâmetro "param"

def runNt(param):
  notebook = WORKER_MAP
  timeout = 30
  args = {"param" : param}
  try:
    result = dbutils.notebook.run(notebook, timeout, args)
    dbutils.notebook.exit("OK")
  except:
    dbutils.notebook.exit(param)
```

## runNtAuto

```
    

#executa a função anterior com recursividade em caso de falha 

def runNtAuto(ORIGIN_LIST = [], DONE_LIST = [], ERROR_LIST = []):
  result = ""
  while True:
    try:
      #SE A ITERAÇÃO DO ÍNDICE _S CORRER SEM ERRO, SERÁ RETORNADO "OK"
      result = runNtLoop(ORIGIN_LIST)
      #CAPTURA-SE O ÍNDICE SEM PROBLEMA 
      output = json.loads(result)
      print(output)
      print("OK")
      # E ADICIONA-SE À LISTA DONE_LIST
      DONE_LIST.append(output)
    except Exception as e:
      
        #SE A ITERAÇÃO DO ÍNDICE _S RESULTAR EM ERRO
        #CAPTURA-SE O ÍNDICE COM PROBLEMA
        output = json.loads(result)
        #ADICIONA-SE À LISTA ERROR_LIST
        ERROR_LIST.append(output)
        #CASO A BASE NÃO TENHA SIDO COMPLETAMENTE ITERADA

        if int(len(DONE_LIST)) + int(len(ERROR_LIST)) < int(len(ORIGIN_LIST)):
        #RETIRA-SE DA LISTA DE ORIGEM OS ÍNDICES JÁ USADOS
          NEXT_ORIGIN_LIST = [x for x in ORIGIN_LIST if x not in DONE_LIST or x not in ERROR_LIST]
          NEXT_DONE_LIST = DONE_LIST
          NEXT_ERROR_LIST = ERROR_LIST
        #REINCIA-SE A MESMA FUNÇÃO COM OS PARÂMETROS AJUSTADOS
    
          runNtAuto( ORIGIN_LIST=NEXT_ORIGIN_LIST,
                         DONE_LIST = NEXT_DONE_LIST,
                         ERROR_LIST =NEXT_ERROR_LIST)
        else:
          result = [NEXT_ORIGIN_LIST, NEXT_DONE_LIST,  NEXT_ERROR_LIST ]
    return result


```

## runNtManual

```
  
def runNtManual(ORIGIN_LIST = [], DONE_LIST = [], ERROR_LIST = []):
  result = ""
  for x in ORIGIN_LIST:
    try:
      #SE A ITERAÇÃO DO ÍNDICE _S CORRER SEM ERRO, SERÁ RETORNADO "OK"
      result = runNtLoop(ORIGIN_LIST)
      print(x)
      print("OK")
      # E ADICIONA-SE À LISTA DONE_LIST
      DONE_LIST.append(x)
    except Exception as e:
        #SE A ITERAÇÃO DO ÍNDICE _S RESULTAR EM ERRO
        #ADICIONA-SE À LISTA ERROR_LIST
        ERROR_LIST.append(x)
        #CASO A BASE NÃO TENHA SIDO COMPLETAMENTE ITERADA

        if int(len(DONE_LIST)) + int(len(ERROR_LIST)) < int(len(ORIGIN_LIST)):
        #RETIRA-SE DA LISTA DE ORIGEM OS ÍNDICES JÁ USADOS
          NEXT_ORIGIN_LIST = [x for x in ORIGIN_LIST if x not in DONE_LIST or x not in ERROR_LIST]
          NEXT_DONE_LIST = DONE_LIST
          NEXT_ERROR_LIST = ERROR_LIST
        #REINCIA-SE A MESMA FUNÇÃO COM OS PARÂMETROS AJUSTADOS
    
          runNtAuto( ORIGIN_LIST=NEXT_ORIGIN_LIST,
                         DONE_LIST = NEXT_DONE_LIST,
                         ERROR_LIST =NEXT_ERROR_LIST)
        else:
          result = [NEXT_ORIGIN_LIST, NEXT_DONE_LIST,  NEXT_ERROR_LIST ]
    return result
  
```
